## 2. Data processing

### 2.1 Delete unused columns

We just need the text without the stopwords it avoid us to make the processing ourself

```python
df = df[['author','language','site_url', 'type','label','title_without_stopwords','text_without_stopwords']]
```

```python
df.head()
```

### 2.2. Regroup all the test in only one column summary

```python
df['summary'] = df['author'] + ' ' + df['title_without_stopwords'] + ' ' + df['text_without_stopwords'] + ' ' + df['language'] + ' ' + df['site_url']
```

```python
# Now delete the columns used for the concatenation
df = df[['summary', 'label']]
```

```python
df.head()
```

### 2.3. Delete the NAs

```python
df = df[df['summary'].notna()]
```

```python
df.isna().sum()
```

### 2.3. Apply binary encoder on label field

```python
encoder = LabelEncoder()
```

```python
df['label'] = encoder.fit_transform(df['label'])
```

```python
df.head()
```

## 3. Save processed dataset

```python
df.to_csv('../../data/processed/001_processed_df.csv', index=False)
```
